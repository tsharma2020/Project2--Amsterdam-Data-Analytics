```python
data_EU_waste = pd.read_csv('../Project2--Amsterdam-Data-Anlytics/data_eu_waste.csv',sep =';')
data_EU_waste.head(5)


#re.sub with apply function
#lambda when calling to get the value of the string into the call to re.sub
```

```python
col_drop = ['Unnamed: 2','Unnamed: 4','Unnamed: 6','Unnamed: 8','Unnamed: 10','Unnamed: 12','Unnamed: 14','Unnamed: 16']
```

```python
#Unit of measure : Tonne
data_EU_waste = data_EU_waste.drop(col_drop, axis = 1)
```

```python
col_drop2b = ['2004', '2006', '2008','2010', '2018']

data_EU_waste = data_EU_waste.drop(col_drop2b, axis = 1)
```
